## load Modules 

```python
import json
import requests

from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer
from konlpy.tag import Mecab
m=Mecab()
```

## load data

```python
with open('/content/drive/MyDrive/gh/test/DATA/sample.json', 'r') as f:
   json_data = json.load(f)
```

```python
articles = json_data['0000001']['articles']

corpus = []
for i in articles:
    corpus.append(i['hilight'])

```

## Tf-idf vectorization

```python
tfidfv = TfidfVectorizer(tokenizer=m.morphs).fit_transform(corpus)
tfidfv.shape
```

## Similarity measurement

```python
document_distances = (tfidfv * tfidfv.T)
document_distances.toarray()[0][0]
```

## Classification

```python
import copy
target = 0
similarity_param = 0.65

scores = document_distances.toarray()
remain_docs = set(list(range(len(corpus))))
classified_docs = []

while True:
    sim_docs = []
    for idx,score in enumerate(scores[target]):
        if score > similarity_param:
            sim_docs.append(idx)

    sim_docs = set.intersection(set(sim_docs), remain_docs)
    classified_docs.append(sorted(list(sim_docs)))

    remain_docs = remain_docs - set(sim_docs)
    if len(list(remain_docs)) == 0:
        break

    target = list(remain_docs)[0]

classified_docs
```

## generate timeline

```python
i = 0
for point in classified_docs:
    idx = point[0]

    if i == 0:
        pass
    else:
        date_after = articles[idx]['published_at']

        if int(date_after) == int(date_before):
            continue
        else:
            print(f">>> 시점 {i}")
            print(articles[idx]['published_at'],articles[idx]['provider'])
            print(articles[idx]['hilight'])
            print("="*100)
            print()
            if i == 16:
                break
    
    date_before = articles[idx]['published_at']
    i += 1
```
