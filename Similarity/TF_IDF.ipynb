## 불용어 처리, 전처리

```python
from konlpy.tag import Mecab
import re
mecab = Mecab()
```

```python
stopwords = []
with open("/content/drive/MyDrive/멀캠/뉴스데이터 해커톤/data/stopwords.txt", encoding='cp949') as f:
    for line in f.readlines():
        stopwords.append(line.strip())
len(stopwords)

stop_pos = ['EC', 'EP', 'SC', 'JK', 'JKO']

preprocessed_docs = []
for key in dict_article.keys():
    words = ''

    re_doc = re.sub('[^가-힣]', ' ', dict_article[key])
    for tag in mecab.pos(re_doc):
        if tag[0] in stopwords or tag[1] in stop_pos:
            continue
        words += tag[0] + ' '

    preprocessed_docs.append(" ".join(mecab.nouns(words[:-1])))
    

print(preprocessed_docs)
```

## TF-IDF 계산

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
count_vect = CountVectorizer(max_df=0.85, max_features=10000)
# count_vect = CountVectorizer()
# 0.85이상인거 뺀다, vocab를 10000개만 만든다.
word_count = count_vect.fit_transform(preprocessed_docs)
print(count_vect.get_feature_names()[:10])

```

```python
import pandas as pd

# overview에서 Null 값을 가진 경우에는 값 제거
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(preprocessed_docs)

# overview에 대해서 tf-idf 수행
print(tfidf_matrix.shape)

```

## 코사인 유사도 구함(문서간 유사도 구하기 위해)

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)# 코사인 유사도 구함
```

```python
# 모든 기사에대해 해당 기사와 유사도 구한다.
sim_scores = list(enumerate(cosine_sim[-1])) 
sim_scores
```

```python
# 유사도에 따라 기사들을 정렬, 맨처음껀 본인.
sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
sim_scores
```

## 핵심 키워드 구하기

```python
def sort_keywords(keywords):
    return sorted(zip(keywords.col, keywords.data), key=lambda x:(x[1], x[0]), reverse=True)
    

def extract_keywords(feature_names, sorted_keywords, n=5):
    """
    전체 키워드에 대한 상위 n개 획득한다.\n
    n에 대한 기본값 5
    """
    return [(feature_names[idx], score) for idx, score in sorted_keywords[:n]]
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer(smooth_idf=True, use_idf=True)
tfidf_transformer.fit(word_count)

for i, key in enumerate(dict_article.keys()):

    doc = preprocessed_docs[i] # 핵심키워드 추출할 문서 선택

    feature_names = count_vect.get_feature_names()
    tfidf_vect = tfidf_transformer.transform(count_vect.transform([doc]))
    sorted_keywords = sort_keywords(tfidf_vect.tocoo())

    # 사용자가 지정한 갯수만큼 키워드 추출
    keywords = extract_keywords(feature_names, sorted_keywords, 5)
    
    print("\n===== 원문 =====")
    print(dict_article[key][:100])
    print("\n=== 핵심키워드 ===")
    for k in keywords:
        print(k)
```
