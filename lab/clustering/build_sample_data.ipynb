```python
def get_attribute(doc):
    url = 'http://tools.kinds.or.kr:8888/feature'

    datas = {
        "access_key": "9af2f705-2974-4340-8b55-d69040b944ab",
        "argument": {
            "title": '', 
            "sub_title": '', 
            "content": doc
            }
    }

    response = requests.post(url, data=json.dumps(datas))

    attribute_scores = response.json()['return_object']['result']['content'].split()
    attribute_keywords = []
    for i in attribute_scores:
        if float(i.split("|")[1]) > 0.037:
            keywords = i.split("|")[0].replace("_", "")
            attribute_keywords.append(keywords)
    
    return attribute_keywords
```
